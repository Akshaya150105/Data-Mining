```python
from collections import defaultdict, namedtuple

class FPNode:
    def __init__(self, item, count, parent):
        self.item = item
        self.count = count
        self.parent = parent
        self.children = {}
        self.link = None

def build_fp_tree(transactions, min_sup):
    item_counts = defaultdict(int)
    for transaction in transactions:
        for item in transaction:
            item_counts[item] += 1

    item_counts = {item: count for item, count in item_counts.items() if count >= min_sup}
    if not item_counts:
        return None, None

    header_table = {item: [count, None] for item, count in item_counts.items()}
    root = FPNode(None, 1, None)

    for transaction in transactions:
        ordered_items = [item for item in sorted(transaction, key=lambda x: item_counts.get(x, 0), reverse=True) if item in item_counts]
        current_node = root
        for item in ordered_items:
            if item in current_node.children:
                current_node.children[item].count += 1
            else:
                new_node = FPNode(item, 1, current_node)
                current_node.children[item] = new_node
                if header_table[item][1] is None:
                    header_table[item][1] = new_node
                else:
                    node = header_table[item][1]
                    while node.link:
                        node = node.link
                    node.link = new_node
            current_node = current_node.children[item]
    return root, header_table

def ascend_fp_tree(node):
    path = []
    while node and node.parent and node.parent.item is not None:
        node = node.parent
        path.append(node.item)
    return path

def find_prefix_paths(base_pat, node):
    cond_pats = {}
    while node:
        path = ascend_fp_tree(node)
        if path:
            cond_pats[frozenset(path)] = node.count
        node = node.link
    return cond_pats

def mine_fp_tree(root,header_table, min_sup, prefix, freq_itemsets):
    sorted_items = [item for item, counts in sorted(header_table.items(), key=lambda p: p[1][0])]
    for base_pat in sorted_items:
        new_freq_set = prefix.copy()
        new_freq_set.add(base_pat)
        freq_itemsets.append((new_freq_set, header_table[base_pat][0]))
        cond_pats = find_prefix_paths(base_pat, header_table[base_pat][1])
        cond_tree, cond_header = build_fp_tree([[item for item in pat] * count for pat, count in cond_pats.items()], min_sup)
        if cond_header:
            mine_fp_tree(cond_tree,cond_header, min_sup, new_freq_set, freq_itemsets)

root, header_table = build_fp_tree(transactions, min_sup)
freq_itemsets_fp = []
if header_table:
    mine_fp_tree(root,header_table, min_sup, set(), freq_itemsets_fp)

print("FP-Growth Frequent Itemsets")
for itemset, support in freq_itemsets_fp:
    print(f"Itemset: {set(itemset)}, Support: {support}")
```

```python
def get_closed_itemsets(frequent_itemsets):
    closed_sets = []
    for itemset, support in frequent_itemsets.items():
        is_closed = True
        for itemset2, support2 in frequent_itemsets.items():
            if itemset!=itemset2 and set(itemset).issubset(itemset2) and support == support2:
                is_closed = False
                break
        if is_closed:
            closed_sets.append((itemset, support))
    return closed_sets
```
